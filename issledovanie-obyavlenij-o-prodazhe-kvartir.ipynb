–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ–º —ç—Ç–∞–∂–∞ –≤ —Ñ–æ—Ä–º–∞—Ç–µ: –ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π, –¥—Ä—É–≥–æ–π –∏ –¥–æ–º (–í—Å–µ–≥–æ —ç—Ç–∞–∂–µ–π = 1 –∏ —ç—Ç–∞–∂ = 1).

````python
first_last = []

for i in range(len(df['floors_total'])):
    if df['floor'][i] == 1:
        if df['floors_total'][i] > 1:
            other = '–ø–µ—Ä–≤—ã–π'
        else:
            if df['floors_total'][i] == 1:
                other = '–¥–æ–º'
                
    elif df['floors_total'][i] == 0:
        other = '–¥—Ä—É–≥–æ–π'
    elif df['floors_total'][i] == df['floor'][i]:
        if df['floor'][i] == df['floors_total'][i]:
            other = '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    else:
        other = '–¥—Ä—É–≥–æ–π'
        
    first_last.append(other)
df['first_last'] = first_last
df.head(5)
````
